```python
import pandas as pd

df = pd.read_csv("clean_heart_failure_clinical_records.csv")

print("Dataset shape:", df.shape)
print("\nColumn names:")
print(df.columns.tolist())

print("\nData types:")
print(df.dtypes)

print("\nMissing values:")
print(df.isnull().sum())

df.columns = df.columns.str.strip().str.lower()

binary_cols = [
    "anaemia",
    "diabetes",
    "high_blood_pressure",
    "sex",
    "smoking",
    "death_event"
]

for col in binary_cols:
    if col in df.columns:
        df[col] = df[col].astype(int)

df.to_csv("formatted_heart_failure_clinical_records.csv", index=False)

print("\nFirst 5 rows:")
print(df.head())

print("\nFormatting complete. File saved as 'formatted_heart_failure_clinical_records.csv'")
```
